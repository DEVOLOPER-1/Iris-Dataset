# **Making a piechart of Species**

```python
counting_species = (iris["species"]).value_counts() #I made this syntax because the plt.pie syntax didn't work because it doesn't accept strings
chosen_labels = ["Iris-setosa","Iris-versicolor","Iris-virginica"]
my_explode = [0.1,0.1,0.1]
chosen_colors = ["#0766AD","#29ADB2","#C5E898"] #These colors were chosen from color hunt palette
plt.pie(counting_species,labels = chosen_labels, colors = chosen_colors, explode = my_explode, shadow = True ,startangle = 140, autopct='%1.1f%%')
#After extensive searching you can put autopct in the pie arguments to put the percentage of each label on the chart
plt.legend(title='Species', bbox_to_anchor=(1, 1))
plt.title("Spechies Piechart")
plt.show
```

## **Making a line graph to illustrate the relationship between “sepal length” and “sepal width”**

```python
bar_sepal_length = iris["sepal_length"]
bar_sepal_width = iris["sepal_width"]
#The values and indecies of sepal lengthes
indecies1 = bar_sepal_length.value_counts().index
values1 = bar_sepal_length.value_counts().values

#The values and indecies of sepal widthes
indecies2 = bar_sepal_width.value_counts().index
values2 = bar_sepal_width.value_counts().values
#Create bar chart

bars = plt.bar(indecies1, values1)
bars2 = plt.bar(indecies2,values2, color = "green")
plt.title("The Relation between Sepal lengthes and widthes by Bar chart")
plt.xlabel=("Indecies")
plt.ylabel=("values") #why the ylabel hasn't  been printed on the graph???
#plt.legend() # Not working as expected
#plt.legend(bars,bars2) # Not working as expected
plt.show()
```

```python
sepal_length = iris["sepal_length"]
sepal_length_indecies = sepal_length.value_counts().index
sepal_legth_values = sepal_length.value_counts().values
plt.plot(sepal_length_indecies,sepal_legth_values,label='Sepal length values', color='red', marker='o')

sepal_width = iris['sepal_width']
sepal_width_indecies = sepal_width.value_counts().index
speal_width_values = sepal_width.value_counts().values
plt.plot(sepal_width_indecies,speal_width_values,label='Sepal width values')

plt.title("The Relation between Sepal lengthes and widthes" ,fontsize=16)
plt.xlabel=("Indecies")
plt.ylabel=("Values")  #why the ylabel hasn't  been printed on the graph???plt.legend()
plt.legend()
plt.grid(True)
plt.show()
```
